```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: better styles
plt.style.use("seaborn-v0_8")
```

```python
# Load the sales dataset
# Try latin1 encoding
df = pd.read_csv(r"C:\Users\Nihal\Desktop\NAIC\Task_3\sales_data_sample.csv", encoding="latin1")

# Convert order date to datetime
df["ORDERDATE"] = pd.to_datetime(df["ORDERDATE"])

# Preview
df.head()
```

```python
# Check dataset structure
df.info()

# Summary statistics
df.describe(include="all")
```

```python
plt.figure(figsize=(12,6))
df.groupby("ORDERDATE")["SALES"].sum().plot(kind="line", marker="o")
plt.title("📈 Total Sales Over Time", fontsize=14)
plt.xlabel("Date")
plt.ylabel("Sales")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x="PRODUCTLINE", y="SALES", data=df, estimator=sum, ci=None, palette="Set2")
plt.title("📊 Sales by Product Line", fontsize=14)
plt.xticks(rotation=45)
plt.show()
```

```python
top_countries = df.groupby("COUNTRY")["SALES"].sum().nlargest(10)

plt.figure(figsize=(12,6))
top_countries.plot(kind="bar", color="skyblue")
plt.title("🌍 Top 10 Countries by Revenue", fontsize=14)
plt.ylabel("Revenue")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.heatmap(df[["QUANTITYORDERED","PRICEEACH","SALES","MSRP"]].corr(), 
            annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("🔥 Correlation Heatmap", fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="DEALSIZE", y="SALES", data=df, palette="Set3")
plt.title("📦 Sales Distribution by Deal Size", fontsize=14)
plt.show()
```
